#### Taking the psychological scores and calculation methods into account to analyse which financial personality type it belongs to

```python
##PSYCHOLOGICAL SCORING KEY
ostrich = [0,0,0,0,'mid','mid','low','low','high','mid','mid','mid','mid','mid','mid','mid','low']
social_val = [0,0,0,0,'mid','mid','high','high','mid','low','low','mid','mid','high','mid','high','low']
fitbit = [0,0,0,0,'high','high','low','mid','mid','mid','high','high','mid','low','mid','mid','high']
anxious_investor = [0,0,0,0,'mid','low','low','mid','mid','low','low','high','mid','high','mid','high','low']
hoarder = [0,0,0,0,'high','mid','low','low','mid','high','mid','mid','high','mid','high','low','high']

ost_score = [0,0,0,0,4,3,3,3,5,3,2,2,2,2,2,2,5]
social_score = [0,0,0,0,3,3,5,5,4,4,0,4,3,1,2,3,5]
fit_score = [0,0,0,0,4,3,4,3,5,5,5,2,2,2,3,2,5]
anx_score = [0,0,0,0,3,4,3,3,3,3,3,5,1,1,3,3,4]
hoarder_score = [0,0,0,0,4,4,1,3,4,4,5,3,2,3,3,2,5]

label = ['Ostrich','Social value spender', 'Fitbit financer', 'Anxious Investor', 'Hoarder']
```
